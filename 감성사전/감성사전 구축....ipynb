```python
#build a model 
model = gensim.models.doc2vec.Doc2Vec(size=50, min_count=2, iter=55) 
model.build_vocab(train_corpus) 
%time model.train(train_corpus, total_examples=model.corpus_count, epochs=model.iter) 
```

```python
with open("p_word.csv") as f:
    reader = csv.reader(f)
    po = [r for r in reader]
    
with open("n_word.csv") as f:
    reader = csv.reader(f)
    ng = [r for r in reader]    
    
#for graph based semi supervised learning
#미리 정의해 놓은 감성사전에 있는 단어를 검색해서 관련있는 단어 top20을 뽑아낸다.
po_data=[]
for i in range(len(po)):

    try: 
        po_data.append(po[i])
        results=model.most_similar(po[i], topn=50)
        for j in results:
            po_data.append([j[0]])
        print('ok')
    except KeyError:
        print("에러")
        

import csv
filename = 'po_data_final' + '.csv'
file = open(filename, 'w', encoding='utf-8', newline = '')
wr = csv.writer(file)
for i in range(0,len(po_data)):
    wr.writerow(po_data[i])

file.close() 



ng_data=list()
for i in range(0,len(ng)):
    try: 
        ng_data.append(ng[i])
        results=model.most_similar(ng[i], topn=50)
        for j in results:
            ng_data.append([j[0]])
        print('ok')
    except KeyError:
        print("에러")

import csv
filename = 'ng_data_final' + '.csv'
file = open(filename, 'w', encoding='utf-8', newline = '')
wr = csv.writer(file)
for i in range(0,len(ng_data)):
    wr.writerow(ng_data[i])

file.close() 
```
