```python
import pandas as pd
df1=pd.read_csv("ID.csv")
df2=pd.read_csv("LN.csv")
df3=pd.read_csv("DLQ.csv")
df4=pd.read_csv("CD_OPN.csv")
```

```python
df4
```

```python
df12=pd.merge(df1,df2)
df12
```

```python
df123=pd.merge(df12,df3,on=('JOIN_KEY','YM',"COM_KEY"),how='left')
df123
```

```python
df1234=pd.merge(df123,df4,on=("YM","JOIN_KEY","COM_KEY"),how="left")
df1234
```

```python
df1234.to_csv("JOIN_F.csv",index=False)
```
